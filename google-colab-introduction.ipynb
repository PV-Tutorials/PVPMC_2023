## Step 2: use pvlib

Here's a simple example of running pvlib code: using pvlib and pandas (a library for data analysis), calculate a clear-sky irradiance profile for Salt Lake City for today (May 10).

```python
import pvlib
import pandas as pd
```

```python
slc = pvlib.location.Location(latitude=40.7608, longitude=-111.8910)
times = pd.date_range(start='2023-05-10 00:00', end='2023-05-10 23:59', freq='5min', tz='US/Mountain')
clearsky_irradiance = slc.get_clearsky(times)
```

```python
clearsky_irradiance.plot(ylabel='Irradiance [W/m2]', title='SLC Clear-Sky Simulation')
```
